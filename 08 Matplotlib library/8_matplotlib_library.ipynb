```python
import numpy as np
import matplotlib.pyplot as plt

x1, y1, score1 = map(float, input("Podaj współczynniki x, y i wynik pierwszego równania: ").split())
x2, y2, score2 = map(float, input("Podaj współczynniki x, y i wynik drugiego równania: ").split())

A = np.array([[x1, y1], [x2, y2]], dtype=float)
B = np.array([score1, score2], dtype=float)

D = np.linalg.det(A)
x_vals = np.linspace(x-10, x+10, 100)

if D != 0:
    A_x = A.copy()
    A_x[:, 0] = B
    D_x = np.linalg.det(A_x)

    A_y = A.copy()
    A_y[:, 1] = B
    D_y = np.linalg.det(A_y)

    x, y = D_x / D, D_y / D

    y_vals1 = (score1 - x1 * x_vals) / y1
    y_vals2 = (score2 - x2 * x_vals) / y2

    plt.plot(x_vals, y_vals1, label=f"{x1}x + {y1}y = {score1}")
    plt.plot(x_vals, y_vals2, label=f"{x2}x + {y2}y = {score2}")
    plt.scatter([x], [y], color='red', label=f"Rozwiązanie ({x:.2f}, {y:.2f})")


else:
    y_vals1 = (score1 - x1 * x_vals) / y1
    y_vals2 = (score2 - x2 * x_vals) / y2
    plt.plot(x_vals, y_vals1, label=f"{x1}x + {y1}y = {score1}")
    plt.plot(x_vals, y_vals2, label=f"{x2}x + {y2}y = {score2}")
    print("Układ jest nieoznaczony lub sprzeczny")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Rozwiązanie układu równań metodą Cramera")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random

rolls_count =[]
dice_value = []
dice_rolls = np.array([(random.randint(1,7)) for _ in range(20)])
print(dice_rolls)

for i in set(dice_rolls):
    dice_value.append(i)
    rolls_count.append(len(np.where(dice_rolls == i)[0]))

plt.bar(dice_value, rolls_count, color="blue")
plt.title("Histogram rzutów kością")
plt.xlabel("Wartość")
plt.ylabel("Liczba wystąpień")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis")

plt.show()
plt.savefig("wykres3D.png")
```

```python
#zad1
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-30, -10, 100)
x2 = np.linspace(-10, 10, 100)
x3 = np.linspace(10, 30, 100)

y1 = 5 * x1 + 131
y2 = x2**2 - 19
y3 = (x3 - 9)**2 + 81

plt.figure(figsize=(8, 6))
plt.plot(x1, y1, label=r'$y = 2x - 1$', color='blue')
plt.plot(x2, y2, label=r'$y = x^2 - 19$', color='green')
plt.plot(x3, y3, label=r'$y = (x-9)^2 + 81$', color='red')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji z trzema przedziałami")
plt.legend()

plt.show()
```

```python
#zad2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 300)  # Od -2π do 2π

y_cos = np.cos(x)
y_sin = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_cos, label=r'$y = \cos(x)$', color='blue')
plt.plot(x, y_sin, label=r'$y = \sin(x)$', color='red')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji y = cos(x) i y = sin(x)")

plt.legend()
plt.show()
```

```python
#zad3
import numpy as np
import matplotlib.pyplot as plt

a = 2
t = np.linspace(0, 15, 100)

v = a * t
s = 0.5 * a * t**2

fig, axs = plt.subplots(2, 1, figsize=(8, 8))

axs[0].plot(t, v, color='blue')
axs[0].set_title('Prędkość w czasie')
axs[0].set_xlabel('Czas (s)')
axs[0].set_ylabel('Prędkość (m/s)')
axs[0].grid(True)

axs[1].plot(t, s, color='red')
axs[1].set_title('Droga w czasie')
axs[1].set_xlabel('Czas (s)')
axs[1].set_ylabel('Droga (m)')

plt.tight_layout()
plt.show()
```
