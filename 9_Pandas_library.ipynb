```python
import pandas as pd

df = pd.read_excel("Gigathon_Wyniki.xlsx")
df
```

```python
print(df.head(3))
print(df.tail(3))
```

```python
df["Wynik procentowy"] = df["Punktacja"] * 100 / 120
df
```

```python
df.dropna(inplace = True)
df
```

```python
df.sort_values(by="Punktacja", ascending=False, inplace=True)
df
```

```python
df["Zakwalifikowany"] = df["Wynik procentowy"] >= 70
df
```

```python
df.groupby(["Zakwalifikowany", "Wiek"])[["Punktacja", "Wynik procentowy"]].agg(["count", "max", "min", "mean", "median"])
```

```python
df_qualified = df[df["Zakwalifikowany"]]
df_qualified.drop("Zakwalifikowany", axis=1, inplace=True)
df_qualified
df_qualified.to_excel("Wyniki_zakwalifikowani.xlsx")
```
